```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# Load the data 
df = pd.read_json('events.json')
# ensure data is ordered
df['event_timestamp'] = pd.to_datetime(df['event_timestamp'])
df = df.sort_values('event_timestamp')
print(df['event_name'].unique())
df.head()
```

```python
# Clean and reformat data
# just sent because "SMS sending peaks"
# foo: a dummy col so that rolling works 
# set index to datetime
df=df[df['event_name'] == 'sent_sms']
df.set_index('event_timestamp', inplace=True)
df['foo'] = 1
```

```python
df['rolling_count'] = df['foo'].rolling(window='1min').count() 
print(df[df['rolling_count']>30].index.unique())
```

```python
df['rolling_count'].plot(figsize=(10, 6))
plt.title('Rolling rate of sent messages/min ')
plt.xlabel('Time')
plt.ylabel('Rate of sent_sms/min')
plt.axhline(y=30, color='r', linestyle='--', label='Warning Rate')
plt.grid(True)
plt.show()
```

```python
# get just intervals 
intervals = []
start= None
end = None
last_end=None
for index, row in df.iterrows():
    if start==None:
        if row['rolling_count']>30:
            start = index
    else:
        if row['rolling_count']<30:
            end=last_end
    if end != None:
        intervals.append([start,end])
        start=None
        end=None
    last_end = index

print(intervals)
```
